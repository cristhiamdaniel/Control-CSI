## Modelo panel fotovoltaico utilizado

El panel utilizado es el Sunset PX-72, que cuenta con 72 celdas de silicio polycristalino. Con la finalidad de obtener una potencia aproximada de 4.073 kW se colocan cuatro paneles en serie y tres en paralelo.

Los parametros utilizados para el primer modelo son:

```tex
T = 25
G = 1000
K_i = 0.037
I_sc = 9.35
T_n = 298
q = 1.6*10**(-19)
E_go = 1.1
n = 1.3
k = 1.38*10**(-23)
R_s = 0.39
R_sh = 545.82
V_oc = 47.4
N_s = 72
NP = 3
NS = 4
```

En donde se consideran las entradas de temperatura e irradiancia como constantes.

### Modelo en Simulink

**Arreglo PV del Sunset PX 72**




**Subsistemas del modelo**



**Foto Corriente**

$I_{ph} = \left(I_{sc} + K_i \times \left(T - 298 \right) \right) \times \left(G/1000 \right)$



**Corriente de saturacion**

$I_o = I_{rs} \times \left( \frac{T}{T_n} \right)^3 \times \exp \left( \left( q \times E_{go} \times \left(1/T_n - 1/T \right) \right)/ \left(n \times k \right) \right)$



**Corriente de saturacion reversa**

$I_{rs} = I_{sc} / (\exp((q*V_{oc})/(n*N_s*k*T ))-1)$



**Corriente shunt**

$I_{sh} = (V +I*R_s)/R_{sh}$



**Corriente de salida**

$I = I_{ph}*NP - I_o * NP *(\exp((q*(V/NS + I*R_s/NP))/(n*N_s*k*T))-1)-I_{sh}$



### Resultados Simulink

**Curva caracteristica VI**



**Curva caracteristica PV**

### Ecuaciones PV

````python
## Parametros
T = Symbol('T')
G = Symbol('G')
K_i = Symbol('K_i')
I_sc = Symbol('I_{sc}')
I_rs = Symbol('I_{rs}')
T_n = Symbol('T_n')
q = Symbol('q')
E_go = Symbol('E_{go}')
n = Symbol('n')
k = Symbol('k')
V = Symbol('V')
I = Symbol('I')
R_s = Symbol('R_s')
R_sh = Symbol('R_{sh}')
V_oc = Symbol('V_{oc}')
N_s = Symbol('N_s')
I_ph = Symbol('I_{ph}')
NP = Symbol('N_P')
I_o = Symbol('I_o')
NS = Symbol('N_S')
I_sh = Symbol('I_{sh}')
````

````python
## Fotocorriente
I_ph = (I_sc + K_i * (T - 298))*(G/1000)
I_ph
````

````python
# Corriente de saturacion
I_o = I_rs * ((T/T_n)**3)*exp((q*E_go*(1/T_n - 1/T))/(n*k))
I_o
````

````python
# Corriente de saturacion reversa
I_rs = I_sc / (exp((q*V_oc)/(n*N_s*k*T ))-1)
I_rs
````

````python
# Corriente Shunt
I_sh = (V +I*R_s)/R_sh
I_sh
````

````python
#Corriente de salida
I = I_ph*NP - I_o * NP *(exp((q*(V/NS + I*R_s/NP))/(n*N_s*k*T))-1)-I_sh
I
````

````python
# parametros del Sunset PX 72

#Funcion de la corriente de salida del arreglo PV
def corrientePV():
    T = 25
    G = 1000
    K_i = 0.037
    I_sc = 9.35
    T_n = 298
    q = 1.6*10**(-19)
    E_go = 1.1
    n = 1.3
    k = 1.38*10**(-23)
    R_s = 0.39
    R_sh = 545.82
    V_oc = 47.4
    N_s = 72
    NP = 3
    NS = 4

    V_pv = Symbol('V_{pv}')
    I_pv = Symbol('I_{pv}')

    I_rs = I_sc / (exp((q*V_oc)/(n*N_s*k*T ))-1)
    I_ph = (I_sc + K_i * (T - 298))*(G/1000)
    I_o = I_rs * ((T/T_n)**3)*exp((q*E_go*(1/T_n - 1/T))/(n*k))
    I_sh = (V_pv +I_pv*R_s)/R_sh
    I_pv =I_ph*NP - I_o * NP *(exp((q*(V_pv/NS + I_pv*R_s/NP))/(n*N_s*k*T))-1)-I_sh
    
    return I_pv

corrientePV()
````

En matlab utilizamos la funcion de ```Piecewise Cubic Hermite Interpolating Polynomial (PCHIP)``` para obtener un vector ```p``` que contenga elementos correspondientes a los elementos de ```vi``` del panel y esten determinados por interpolacin cbica por partes dentro de los vectores ```v``` e ```i```.

```matlab

ylim([0 30]);
xlswrite('pchip', [x, y])
```

````python
# cargamos los datos del simulink en un dataframe
df2 = pd.read_csv("pchip.csv", header=None)
df2.columns = ['v','p']
df2.head()
````

````python
plt.scatter(df2['v'], df2['p'])
plt.grid()
plt.show()
````

**Ajustamos la curva**

````python
#fit polynomial models up to degree 5
model1 = np.poly1d(np.polyfit(df2.v, df2.p, 1))
model2 = np.poly1d(np.polyfit(df2.v, df2.p, 2))
model3 = np.poly1d(np.polyfit(df2.v, df2.p, 3))
model4 = np.poly1d(np.polyfit(df2.v, df2.p, 4))
model5 = np.poly1d(np.polyfit(df2.v, df2.p, 5))

#create scatterplot
polyline = np.linspace(1, 200, 100)
plt.scatter(df2.v, df2.p)

#add fitted polynomial lines to scatterplot 
plt.plot(polyline, model1(polyline), color='green')
plt.plot(polyline, model2(polyline), color='red')
plt.plot(polyline, model3(polyline), color='purple')
plt.plot(polyline, model4(polyline), color='blue')
plt.plot(polyline, model5(polyline), color='orange')
plt.show()
````

````python
#define function to calculate adjusted r-squared
def adjR(x, y, degree):
    results = {}
    coeffs = np.polyfit(x, y, degree)
    p = np.poly1d(coeffs)
    yhat = p(x)
    ybar = np.sum(y)/len(y)
    ssreg = np.sum((yhat-ybar)**2)
    sstot = np.sum((y - ybar)**2)
    results['r_squared'] = 1- (((1-(ssreg/sstot))*(len(y)-1))/(len(y)-degree-1))

    return results

#calculated adjusted R-squared of each model
print(adjR(df2.v, df2.p, 1))
print(adjR(df2.v, df2.p, 2))
print(adjR(df2.v, df2.p, 3))
print(adjR(df2.v, df2.p, 4))
print(adjR(df2.v, df2.p, 5))
````

````python
#fit fourth-degree polynomial
model5 = np.poly1d(np.polyfit(df2.v, df2.p, 5))

#define scatterplot
polyline = np.linspace(1, 200, 100)
plt.scatter(df2.v, df2.p)

#add fitted polynomial curve to scatterplot
plt.plot(polyline, model5(polyline), '--', color='red')
plt.show()
````
